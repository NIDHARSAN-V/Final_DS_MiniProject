## Fixing Datatypes and standardising

```python
df_validation = pd.read_csv("../../dataset/01_raw/billings/data_validation_summary.csv")
print(df_validation.info)
```

```python
df_mixed = pd.read_csv("../../dataset/01_raw/billings/mixed_data_types.csv")
print(df_mixed)
```

```python
df_billing['Connection_Group'].unique()
```

```python
df_billing['Connection_Group'] = df_billing['Connection_Group'].replace('nan', 'unknown')
df_billing['Connection_Group'] = df_billing['Connection_Group'].str.strip().str.lower()
df_billing['Connection_Group'] = df_billing ['Connection_Group'].astype(str)
df_billing['Connection_Group'] = df_billing['Connection_Group'].replace({
    '4 to 9': '4-9',
    '10+': '10+',
    'independent': 'independent',
    'unknown': 'unknown'
})
```

```python
df_billing['Connection_Group'].unique()
```

```python
df_billing['Tenure_Group'].unique()
```

```python
df_billing['Tenure_Group'] = df_billing['Tenure_Group'].replace('nan', 'unknown')

df_billing['Tenure_Group'] = df_billing['Tenure_Group'].str.strip().str.lower()

df_billing['Tenure_Group'] = df_billing ['Tenure_Group'].astype(str)

df_billing['Tenure_Group'] = df_billing['Tenure_Group'].replace({
    '4+': '4+',
    'unknown': 'unknown',
    
})
```

```python
df_billing['Tenure_Group'].unique()
```

```python
df_billing['Anchor_Group'].unique()
```

```python
df_billing['Anchor_Group'] = df_billing['Anchor_Group'].replace('nan', 'unknown')

df_billing['Anchor_Group'] = df_billing['Anchor_Group'].str.strip().str.lower()

df_billing['Anchor_Group'] = df_billing ['Anchor_Group'].astype(str)
df_billing['Anchor_Group'] = df_billing['Anchor_Group'].replace({
    '4 to 9': '4-9',
    '10+': '10+',
    'independent': 'independent',
    'unknown': 'unknown'
})
```

```python
df_billing['Anchor_Group'].unique()
```

```python
df_billing['Discount_Amount'] = df_billing['Discount_Amount'].fillna(0)
```

## Analyzing Cols with more NULLs 

```python
print(len(df_billing))
# print(df_billing.isnull().sum())

print(len(df_billing) - df_billing.isnull().sum())
```

```python
print(
    df_validation[['column_name', 'column_type', 'null_percentage']]
    .sort_values(by='null_percentage', ascending=False)
)
```

```python
print("Before dropping columns:", len(df_billing.columns))

df_billing = df_billing.drop(
    columns=df_validation[df_validation['null_percentage'] > 90]['column_name'].tolist(),
    errors='ignore'
)
print("After dropping columns:", len(df_billing.columns))

```
