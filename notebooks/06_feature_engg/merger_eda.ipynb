# Dropping Features basic 

```python
drop_cols = ['Co_Ref', 'Call_ID_renewal', 'Contact_ID_cc']
date_cols = [
    'Proforma_Date',
    'Registration_Date',
    'Prospect_Renewal_Date',
    'Closed_Date',
    'Last_Renewal',
    'DateTime_Out',
    'Prev_Renewal_Date',
    'Call_Date_renewal',
    'Call_Date_cc'
]
leakage_cols = [
    'Prospect_Outcome',   # directly linked to target
    'Closed_Date',        # happens after outcome
    'Renewal_Year',       # derived from outcome timeline
    'Last_Renewal',       # future info depending on context
]
```

```python
df['days_to_renewal'] = (df['Prospect_Renewal_Date'] - df['Call_Date_renewal']).dt.days

df['customer_tenure_days'] = (df['Prospect_Renewal_Date'] - df['Registration_Date']).dt.days

df['days_since_last_renewal'] = (df['Prospect_Renewal_Date'] - df['Prev_Renewal_Date']).dt.days

df['last_cc_call_gap'] = (df['Prospect_Renewal_Date'] - df['Call_Date_cc']).dt.days
```

```python
df.drop(columns=date_cols, inplace=True, errors='ignore')
```

```python
ct = pd.crosstab(df['days_to_renewal'], df['target'])

```

```python
import matplotlib.pyplot as plt

features = [
    'days_to_renewal',
    'customer_tenure_days',
    'days_since_last_renewal',
    'last_cc_call_gap'
]

for col in features:
    plt.figure()

    df[df['target'] == 0][col].plot(kind='hist', bins=50, alpha=0.5, label='No Churn (0)')
    df[df['target'] == 1][col].plot(kind='hist', bins=50, alpha=0.5, label='Churn (1)')

    plt.title(f'{col} Distribution by Target')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()
```
